```python
# Padronizaçã de valores categóricos
country_mapping = {
    "United-States": "United States",
    "Outlying-US(Guam-USVI-etc)": "United States",
    "Holand-Netherlands": "Netherlands",
    "Trinadad&Tobago": "Trinidad & Tobago",
    "Hong": "Hong Kong",
}
full_df['native_country'] = full_df['native_country'].replace(country_mapping)

full_df['native_country'].value_counts()
```

```python
# Converte valore ? para ausestes
full_df.replace('?', np.nan, inplace=True)
```

```python
# Colunas numericas para normalização
numeric_cols = full_df.select_dtypes(include=['int64', 'float64']).columns
df_numerico = full_df[numeric_cols].copy() # Copia do dataset original
```

```python
# Min e Max Scaler
minmax_scaler = MinMaxScaler()
df_minmax = pd.DataFrame(
    minmax_scaler.fit_transform(df_numerico),
    columns=numeric_cols
)
```

```python
# Z-Score
standard_scaler = StandardScaler()
df_standard = pd.DataFrame(
    standard_scaler.fit_transform(df_numerico),
    columns=numeric_cols
)
```

```python
# Robust Scaler
robust_scaler = RobustScaler()
df_robust = pd.DataFrame(
    robust_scaler.fit_transform(df_numerico),
    columns=numeric_cols
)
```

```python
# Comparação das normalizações
comparacao = pd.DataFrame({
    "Original Média": df_numerico.mean(),
    "Min-Max Média": df_minmax.mean(),
    "Z-Score Média": df_standard.mean(),
    "Robust Média": df_robust.mean()
})

comparacao
```
