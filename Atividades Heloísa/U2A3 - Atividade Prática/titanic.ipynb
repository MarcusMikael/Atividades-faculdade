```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
```

```python
# Analise inicial
print(df.head())
```

```python
# Informações gerais do dataset
print(df.info())
```

```python
# Valores duplicados
print(df.duplicated().sum())
```

Não possui valores duplicados nesse dataset.

```python
# Valores ausentes
print(df.isnull().sum())
```

Possui alguns valores ausentes em 3 colunas.

```python
df.columns
```

```python
len(df['Age'])
```

```python
# Para a coluna age(idade) irei usar a mediana para a substituição dos valores ausentes, pois a mediana evita distorções causadas por idades muito extremas.
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
print('Quantidade de dados na coluna Cabin:')
len(df['Cabin'])
```

```python
print('Quantidade de valores ausentes:')
df['Cabin'].isnull().sum()
```

Na coluna Cabin, a maioria dos dados estão ausentes, então vou removê-la por essa questão.

```python
# Remoção da coluna cabin
df.drop(columns=['Cabin'], inplace=True)
```

```python
len(df['Embarked'])
```

```python
# Coluna Embarked: vou preencher com a moda, pois são apenas dois valores ausentes
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
# Descrição estatísticas
print(df.describe())
```

```python
# Tipos de dados
print(df.dtypes)
```

```python
# Padronizar os dados
# Features
# Tranforma a coluna sex em valores numericos
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

As análises são melhores com dados numéricos

```python
# Embarked: One hot encoding
df = pd.get_dummies(df, columns=['Embarked'], prefix='Emb')
# evitar atribuir ordens falsas em categorias nominais.
# Isso transforma a coluna Embarked em várias colunas. Sendo representado por 0 e 1
```
